```python
"""En la lecci´on del 24 de setiembre del 2024, se explic´o la aplicaci´on del efecto croma en im´agenes.
Implementar en Python, el efecto croma estudiado en clase, basado en el c´odigo generado en GNU
Octave. Para eso, utilice las im´agenes fondo verde.jpg y playa.jpg, que se encuentran en la
carpeta de One Drive de dicha lecci´on"""

import cv2 as cv
from matplotlib import pyplot as pl
import numpy as np
from scipy.ndimage import binary_erosion
from scipy.signal import convolve2d


'''def borde(A):
    # Crear una máscara binaria basada en las condiciones dadas
    B = (A[:, :, 0] < 100) & (A[:, :, 1] >= 125) & (A[:, :, 2] < 100)
    
    # Crear un elemento estructurante (matriz de unos de 3x3)
    M = np.ones((3, 3), dtype=bool)
    
    # Realizar la erosión binaria
    C1 = binary_erosion(B, structure=M).astype(np.uint8)
    
    # Restar la imagen erosionada de la imagen binaria original
    C = np.uint8(B & ~C1) * 255
    
    # Mostrar el resultado usando matplotlib
    pl.figure("ventana 2")
    pl.imshow(C, cmap='gray')
    pl.axis('off')  # Para ocultar los ejes
    pl.show()
    return C '''
# Define the border detection function
def borde(A):
    # Identify greenish pixels in the image
    B = (A[:, :, 0] < 100) & (A[:, :, 1] >= 125) & (A[:, :, 2] < 100)
    
    # Dilate the border
    M = np.ones((3, 3), np.uint8)
    C1 = cv.dilate(B.astype(np.uint8), M)
    
    # Create the external border by subtracting the original from the dilated
    borde_dilatado = (C1 > 0) & (~B)

    return (borde_dilatado * 255).astype(np.uint8)


def fondo_verde(A,B):
    #Asegurarse de que ambas imagenes son iguales
   if(A.shape!=B.shape):
      raise("las imagenes A y B deben de tener el mismo tamaño")
   
    # Crear imágenes vacías para los resultados intermedios
   C1 = np.zeros_like(A, dtype=np.uint8)  # Imagen A con fondo negro en lugar de verde
   C2 = np.zeros_like(A, dtype=np.uint8)  # Imagen B con negro en los píxeles de la imagen que no es verde de A

    #  # Definir máscara para detectar el verde (ajustar tolerancia si es necesario)
   #mask1 = (A[:, :, 0] < 100) & (A[:, :, 1] >= 125) & (A[:, :, 2] < 100)
   mask1 = (A[:, :, 0] < 100) & (A[:, :, 1] >= 100) & (A[:, :, 2] < 100)

   mask2 = np.logical_not(mask1)
   # Obtener persona de la imagen A, con fondo negro
   C1[:, :, 0] = mask2 * A[:, :, 0]
   C1[:, :, 1] = mask2 * A[:, :, 1]
   C1[:, :, 2] = mask2 * A[:, :, 2]

   # Obtener fondo de la imagen B, pero negro en la posición de la persona en A
   C2[:, :, 0] = mask1 * B[:, :, 0]
   C2[:, :, 1] = mask1 * B[:, :, 1]
   C2[:, :, 2] = mask1 * B[:, :, 2]

    # Combinar las dos imágenes
   C = C1 + C2

   return C




def conv2_borde(imagen, mascara, iteraciones=1000):
    """
    Aplica difusión isotrópica a una imagen usando una máscara.

    Args:
    imagen (numpy.ndarray): Imagen original (en formato RGB).
    mascara (numpy.ndarray): Máscara binaria (donde 255 son los píxeles a actualizar).
    iteraciones (int): Número de iteraciones para la difusión.

    Returns:
    numpy.ndarray: Imagen restaurada.
    """
    # Definir el kernel de difusión isotrópica
    kernel = np.array([[0.073235, 0.176765, 0.073235],
                       [0.176765, 0.0, 0.176765],
                       [0.073235, 0.176765, 0.073235]])

    # Crear una copia de la imagen a restaurar
    imagen_restaurada = imagen.copy()

    # Realizar el proceso de difusión isotrópica
    for iter in range(iteraciones):
        for canal in range(imagen.shape[2]):  # Iterar sobre los canales (R, G, B)
            # Convolucionar toda la imagen
            convolucion = convolve2d(imagen_restaurada[:, :, canal], kernel, mode='same')
            
            # Actualizar solo los píxeles afectados por la marca (mascara == 255)
            imagen_restaurada[:, :, canal] = np.where(mascara == 255, convolucion, imagen[:, :, canal])

    return imagen_restaurada


A=cv.imread("imagenes/fondo_verde.jpg")
B=cv.imread("imagenes/playa.jpg")
A=cv.cvtColor(A, cv.COLOR_BGR2RGB)
B=cv.cvtColor(B, cv.COLOR_BGR2RGB)
C=fondo_verde(A,B)

pl.subplot(2,2,1)
pl.imshow(A)




pl.subplot(2,2,2)
pl.imshow(B)





#pl.subplot(2,2,3)
pl.figure("contorno")
pl.imshow(C)

#imagen con solo el borde de la persona
D=borde(A)



E=conv2_borde(C,D,10)
pl.figure()
pl.imshow(E)



pl.tight_layout()
pl.show()
```

<h3>  
    En la lección del 26 de setiembre del 2024, se explicó la transformada de Hough para detectar lineas
rectas en una imagen. Implemente computacionalmente en Python el archivo ejemplo2.m, para
detectar las lineas rectas de la imagen linea2.jpg.
 </h3>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
A = cv2.imread('imagenes/linea2.jpg', cv2.IMREAD_GRAYSCALE)

# Convert to binary
B = np.double(A)
B[B < 0.5 * 255] = 0  # Since B is in range [0, 255]
B[B >= 0.5 * 255] = 1
plt.imshow(B, cmap='gray')

# Discretization of angle and p
h1 = 0.5  # Step between angles
angles = np.deg2rad(np.arange(0, 180 + h1, h1))
m, n = B.shape
d = np.sqrt(m ** 2 + n ** 2)
h2 = 0.5  # Step between p
ps = np.arange(-d, d + h2, h2)

# Create the accumulation matrix
m1 = len(angles)
n1 = len(ps)
Accumulator = np.zeros((m1, n1))

# Fill the accumulation matrix
x_b, y_b = np.where(B == 1)  # Coordinates where edges are detected

for i in range(len(x_b)):
    for ang_ind in range(len(angles)):
        ang = angles[ang_ind]
        p = x_b[i] * np.cos(ang) + y_b[i] * np.sin(ang)
        p_ind = np.argmin(np.abs(ps - p))
        Accumulator[ang_ind, p_ind] += 1

# Plot the behavior of the accumulator matrix (optional)
# plt.figure()
# plt.imshow(Accumulator, extent=[np.rad2deg(angles[0]), np.rad2deg(angles[-1]), ps[0], ps[-1]], aspect='auto')
# plt.xlabel('Angles (degrees)')
# plt.ylabel('p values')
# plt.colorbar()

# Number of lines to detect
num_lines = 6

for i in range(num_lines):
    # Find the position where the Accumulator reaches its maximum
    maxA = np.max(Accumulator)
    xp, yp = np.where(Accumulator == maxA)

    for k in range(len(xp)):
        # Draw the line on the original image
        angMax = angles[xp[k]]
        pMax = ps[yp[k]]

        if abs(np.sin(angMax)) < 1e-5:
            x_v = pMax / np.cos(angMax)
            plt.plot([0, n], [x_v, x_v], 'r', linewidth=2)
            plt.xlim([0, n])
            plt.ylim([m, 0])
        else:
            # Calculate slope and intercept
            slope = -np.cos(angMax) / np.sin(angMax)
            intercept = pMax / np.sin(angMax)
            # Calculate points (1, y1), (m, ym), (x1, 1), and (xn, n)
            y1 = slope * 1 + intercept
            ym = slope * m + intercept
            x1 = (1 - intercept) / slope
            xn = (n - intercept) / slope

            if slope > 0:
                if y1 >= 0:
                    plt.plot([1, xn], [y1, n], 'r', linewidth=2)
                else:
                    plt.plot([1, ym], [x1, m], 'r', linewidth=2)
            else:
                if y1 > m:
                    plt.plot([ym, n], [m, xn], 'r', linewidth=2)  # Corrected from xm to ym
                else:
                    plt.plot([y1, 1], [1, x1], 'r', linewidth=2)

            plt.xlim([0, n])
            plt.ylim([m, 0])

        # Reset accumulator maximum to avoid redetection
        Accumulator[xp[k], yp[k]] = 0

# Show the result
plt.show()

```

<h3>  La lección del 1° de octubre del 2024, se inicio el tema de segmentación de imágenes. Implemente computacionalmente en Python el método iterativo del umbral global ´optimo, y aplique dicho
algoritmo para segmentar la imagen imagen5.jpg.
 </h3>

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
# Cargar la imagen
A = cv2.imread('imagenes/imagen5.jpg', cv2.IMREAD_GRAYSCALE)
plt.subplot(2, 2, 1)
plt.imshow(A, cmap='gray')
plt.title('Imagen Original')
# Dimensiones de la imagen
m, n = A.shape
# Histograma de la imagen original
plt.subplot(2, 2, 2)
plt.hist(A.ravel(), bins=256, range=(0, 256))
plt.title('Histograma de la Imagen Original')
# Algoritmo del umbral básico
T = 165  # Paso 1
iterMax = 1000
tol = 1e-5
for k in range(iterMax):  # Paso 2
    # Paso 3
    I1 = (A > T)  # Matriz binaria donde será 1 cuando A > T, y 0 cuando A <= T
    I2 = (A <= T)  # Matriz binaria donde será 1 cuando A <= T, y 0 cuando A > T

    B1 = A * I1  # Bloque 1
    B2 = A * I2  # Bloque 2

    m1 = np.sum(B1) / np.sum(I1)  # Promedio de intensidad del Bloque 1
    m2 = np.sum(B2) / np.sum(I2)  # Promedio de intensidad del Bloque 2

    Tn = (m1 + m2) / 2
    er = abs(Tn - T)
    
    if er < tol:
        break
    
    T = Tn
# Crear la imagen binaria
C = np.zeros((m, n), dtype=np.uint8)
C[A > T] = 1
C[A <= T] = 0
# Mostrar el resultado
plt.subplot(2, 2, 3)
plt.imshow(C, cmap='gray')
plt.show()

```

explicó el método zig-zag, el cual se define a continuación:
 </h3>

```python
function x = mat2vect_diag(A)

    % Inicializar el vector de salida
    x = [];

    % Recorrer las diagonales de la matriz en zig-zag
    for s = 1:(size(A, 1) + size(A, 2) - 1)
        if mod(s, 2) == 1  % Diagonales de abajo hacia arriba
            for j = 1:min(s, size(A, 2))
                i = s - j + 1;  %  índice de fila
                if i >= 1 && i <= size(A, 1) && j >= 1 && j <= size(A, 2)
                    if A(i, j) ~= 0  % Solo agregar si el elemento es diferente de cero
                        x(end + 1) = A(i, j);
                    end
                end
            end
        else  % Diagonales de arriba hacia abajo
            for i = 1:min(s, size(A, 1))
                j = s - i + 1;  %  índice de columna
                if i >= 1 && i <= size(A, 1) && j >= 1 && j <= size(A, 2)
                    if A(i, j) ~= 0  % Solo agregar si el elemento es diferente de cero
                        x(end + 1) = A(i, j);
                    end
                end
            end
        end
    end
end

function C = vect2mat_diag(x)


    % Variable para mantener el índice del vector x
    idx = 1;

    % Recorrer las diagonales de la matriz
    for s = 1:(size(C, 1) + size(C, 2) - 1)
        if mod(s, 2) == 1  % Diagonales de abajo hacia arriba
            for j = 1:min(s, size(C, 2))
                i = s - j + 1;  % índice de fila
                if i >= 1 && i <= size(C, 1) && j >= 1 && j <= size(C, 2) && idx <= length(x)
                    C(i, j) = x(idx);
                    idx = idx + 1;
                end
            end
        else  % Diagonales de arriba hacia abajo
            for i = 1:min(s, size(C, 1))
                j = s - i + 1;  % índice de columna
                if i >= 1 && i <= size(C, 1) && j >= 1 && j <= size(C, 2) && idx <= length(x)
                    C(i, j) = x(idx);
                    idx = idx + 1;
                end
            end
        end
    end
end




A=[ 1    9   17   25   33   41   49   57;
    2   10   18   26   34   42   50   58;
    3   11   19   27   35   43   51   0;
    4   12   20   28   36   44    0   0;
    5   13   21   29   37    0    0   0;
    6   14   22   30    0    0    0   0;
    7   15   23    0    0    0    0   0;
    8   16    0    0    0    0    0   0]

x=mat2vect_diag(A)

C=vect2mat_diag(x)
```

<h3>En la lección del 17 de octubre, se explicó la aplicación de la SVD en el tema de compresión de
imágenes y se implemento un ejemplo en GNU Octave. Implemente computacionalmente en Python
el ejemplo de compresión de imágenes que se encuentra en el archivo ejemplo3.m. Para eso, tiene
que implementar adicionalmente las funciones rango reducido y r optimo95</h3>

```python
import numpy as np

def rango_reducido(A, r):
    # Check if r is valid
    if r > min(A.shape):
        raise ValueError("r must be less than or equal to the dimensions of matrix A")
    
   
    U, S, Vt = np.linalg.svd(A)
    
    
    Ur = U[:, :r]    
    Sr = np.diag(S[:r])  
    Vr = Vt[:r, :]   
    

    Xr = Ur @ Sr     
    Yr = Vr        

    return Xr, Yr
def r_optimo95(A):
   
    s = np.linalg.svd(A, compute_uv=False)
    p = len(s)
    
  
    for r in range(1, p + 1):
        cr = np.sum(s[:r]) / np.sum(s)
        if cr > 0.95:
            break
    
    return r


import matplotlib.pyplot as plt
import cv2



A = cv2.imread('imagenes/img1.jpg', cv2.IMREAD_GRAYSCALE) / 255.0

# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(A, cmap='gray')
plt.title('Original Image')
plt.axis('off')


np.savetxt('originalImage.txt', A)


m, n = A.shape
print(f"Number of elements in A = {m * n}")


r = r_optimo95(A)


Xr, Yr = rango_reducido(A, r)
m1, n1 = Xr.shape
m2, n2 = Yr.shape
print(f"Number of elements stored in Xr and Yr = {m1 * n1 + m2 * n2}")


np.savetxt('matrixXr.txt', Xr)
np.savetxt('matrixYr.txt', Yr)


Ar = np.dot(Xr, Yr)


compression_ratio = (m * r + r * n) / (m * n)
print(f"Compression Ratio of Ar = {compression_ratio}")


plt.subplot(1, 2, 2)
plt.imshow(Ar, cmap='gray')
plt.title('Imagen reconstruida')
plt.axis('off')


plt.figure()
s = np.linalg.svd(A, compute_uv=False)
plt.plot(range(1, len(s) + 1), s)
plt.xlabel('index de valor singular')
plt.ylabel('valor singular')
plt.title('valores singulares de A')
plt.show()

```

<h1>Ejercicio 9</h1>
<h3>En la lección del 22 de octubre, se explicó la aplicación de la SVD en el tema de eliminación de
ruidos en GNU Octave. Implemente computacionalmente en Python el ejemplo de eliminación de
ruido que se encuentra en el archivo ejemplo3.m</h3>


```python
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt

# Parameters
folder = 'imagen_nasa_saturno'
sigma = 0.1

# Load the first image to get dimensions
A1 = cv2.imread(os.path.join(folder, 'saturno (1).jpg'), cv2.IMREAD_GRAYSCALE)
m, n = A1.shape

# Get a list of all images in the folder
image_files = [f for f in os.listdir(folder) if f.endswith('.jpg')]
num_img = len(image_files)

# Initialize matrices A and B for training images
A = np.zeros((m * n, num_img))
B = np.zeros((m * n, num_img))

# Obtain matrices A and B for training
for i in range(num_img):
    # Read and convert image to double precision (float64)
    C = cv2.imread(os.path.join(folder, f'saturno ({i+1}).jpg'), cv2.IMREAD_GRAYSCALE)
    C = C.astype(np.float64) / 255.0  # Normalize image to [0, 1]
    A[:, i] = C.flatten()  # Flatten image into a vector

    # Add noise to the image
    R = C + sigma * np.random.randn(m, n)
    B[:, i] = R.flatten()

# Compute the Super Filter
F = A @ np.linalg.pinv(B)

# Clean a new noisy image
Xnew = cv2.imread('imagenes/imgN.jpg', cv2.IMREAD_GRAYSCALE).astype(np.float64) / 255.0  # Load original image
plt.subplot(2, 2, 1)
plt.imshow(Xnew, cmap='gray')
plt.title('Original Image')

# Add noise to the image
Ynew = Xnew + sigma * np.random.randn(m, n)
plt.subplot(2, 2, 2)
plt.imshow(Ynew, cmap='gray')
plt.title('Noisy Image')

# Vectorize the noisy image and apply the filter
bt = Ynew.flatten()
at = F @ bt

# Reshape the cleaned vector back into an image
Xrec = at.reshape(m, n)
plt.subplot(2, 2, 3)
plt.imshow(Xrec, cmap='gray')
plt.title('Reconstructed Image')

# Show plots
plt.tight_layout()
plt.show()
```
